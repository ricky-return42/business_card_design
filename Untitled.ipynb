```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
import itertools
%matplotlib inline
sns.set_style('white')

X = load_iris().data
y = load_iris().target
```

```python
df = pd.DataFrame(X,y)
df = df.reset_index()
```

```python
gp0 = df[df.iloc[:,0]==0]
gp1 = df[df.iloc[:,0]==1]
gp2 = df[df.iloc[:,0]==2]
```

```python
fig = plt.figure(figsize=(10,21))
ax1 = fig.add_subplot(311)
ax2 = fig.add_subplot(312)
ax3 = fig.add_subplot(313)
ax1.scatter(gp0.iloc[:,3], gp0.iloc[:,4], c='r', alpha=.35)
ax1.scatter(gp1.iloc[:,3], gp1.iloc[:,4], c='darkgrey', alpha=.5)
ax1.scatter(gp2.iloc[:,3], gp2.iloc[:,4], c='dimgrey', alpha=.5)
ax1.patch.set_facecolor('#00062E')

ax2.scatter(gp0.iloc[:,3], gp0.iloc[:,4], c='dimgrey', alpha=.5)
ax2.scatter(gp1.iloc[:,3], gp1.iloc[:,4], c='b', alpha=.6)
ax2.scatter(gp2.iloc[:,3], gp2.iloc[:,4], c='darkgrey', alpha=.5)
ax2.patch.set_facecolor('#00062E')

ax3.scatter(gp0.iloc[:,3], gp0.iloc[:,4], c='darkgrey', alpha=.5)
ax3.scatter(gp1.iloc[:,3], gp1.iloc[:,4], c='dimgrey', alpha=.5)
ax3.scatter(gp2.iloc[:,3], gp2.iloc[:,4], c='g', alpha=.6)
ax3.patch.set_facecolor('#00062E')

plt.show()
```

```python
fig = plt.figure(figsize=(10,21))
ax1 = fig.add_subplot(311)
ax2 = fig.add_subplot(312)
ax3 = fig.add_subplot(313)
ax1.scatter(gp0.iloc[:,1], gp0.iloc[:,2], c='r', alpha=.35)
ax1.scatter(gp1.iloc[:,1], gp1.iloc[:,2], c='darkgrey', alpha=.5)
ax1.scatter(gp2.iloc[:,1], gp2.iloc[:,2], c='dimgrey', alpha=.5)
ax1.patch.set_facecolor('#00062E')

ax2.scatter(gp0.iloc[:,1], gp0.iloc[:,2], c='dimgrey', alpha=.5)
ax2.scatter(gp1.iloc[:,1], gp1.iloc[:,2], c='b', alpha=.6)
ax2.scatter(gp2.iloc[:,1], gp2.iloc[:,2], c='darkgrey', alpha=.5)
ax2.patch.set_facecolor('#00062E')

ax3.scatter(gp0.iloc[:,1], gp0.iloc[:,2], c='darkgrey', alpha=.5)
ax3.scatter(gp1.iloc[:,1], gp1.iloc[:,2], c='dimgrey', alpha=.5)
ax3.scatter(gp2.iloc[:,1], gp2.iloc[:,2], c='g', alpha=.6)
ax3.patch.set_facecolor('#00062E')

plt.show()
```

```python
itlst = list(itertools.combinations([1,2,3,4],2))
```

```python
for (x,y) in itlst:
    fig = plt.figure(figsize=(15.5,3))
    ax1 = fig.add_subplot(131)
    ax2 = fig.add_subplot(132)
    ax3 = fig.add_subplot(133)
    ax1.scatter(gp0.iloc[:,x], gp0.iloc[:,y], c='crimson', alpha=.4, s=30)
    ax1.scatter(gp1.iloc[:,x], gp1.iloc[:,y], c='lightgrey', alpha=.5, s=30)
    ax1.scatter(gp2.iloc[:,x], gp2.iloc[:,y], c='dimgrey', alpha=.5, s=30)
    ax1.patch.set_facecolor('#000525')

    ax2.scatter(gp0.iloc[:,x], gp0.iloc[:,y], c='dimgrey', alpha=.5, s=30)
    ax2.scatter(gp1.iloc[:,x], gp1.iloc[:,y], c='dodgerblue', alpha=.6, s=30)
    ax2.scatter(gp2.iloc[:,x], gp2.iloc[:,y], c='lightgrey', alpha=.5, s=30)
    ax2.patch.set_facecolor('#000525')

    ax3.scatter(gp0.iloc[:,x], gp0.iloc[:,y], c='lightgrey', alpha=.5, s=30)
    ax3.scatter(gp1.iloc[:,x], gp1.iloc[:,y], c='dimgrey', alpha=.5, s=30)
    ax3.scatter(gp2.iloc[:,x], gp2.iloc[:,y], c='springgreen', alpha=.6, s=30)
    ax3.patch.set_facecolor('#000525')

    plt.show()
```
